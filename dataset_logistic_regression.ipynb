```python
import numpy as np

def sigmoid(x):
  return (1/(1 + np.exp(-x)))
```

```python
theta_0 = 0.0
theta_1 = 0.0

def hypothesis(x):
  return sigmoid(theta_0 + theta_1*x);
```

```python
def gradient_descent(x, y, m, iter, alpha):
  global theta_0,theta_1
  h = 0.0
  e = 0.0
  sum_e = 0.0
  e_mul_x = 0.0
  sum_e_mul_x = 0.0

  for i in range(iter):
    h = 0.0
    e = 0.0
    sum_e = 0.0
    e_mul_x = 0.0
    sum_e_mul_x = 0.0
    
    for j in range(m):
      # Calculate hypothesis
      print(x[j])
      h = hypothesis(x[j]);
      # Calculate error
      e = h - y[j];
      # Calculate sum of error
      sum_e = sum_e + e
      # Calculate error*x
      e_mul_x = e * x[j]
      # Calculate sum of error*x
      sum_e_mul_x = sum_e_mul_x + e_mul_x
    

    # Calculate d_J/d_theta_0
    d_J_d_theta_0 = sum_e / m;  
    # Calculate d_J/d_theta_1
    d_J_d_theta_1 = sum_e_mul_x / m;
    
    # Update theta 0
    theta_0 = theta_0 - alpha*d_J_d_theta_0;    
    # Update theta 1
    theta_1 = theta_1 - alpha*d_J_d_theta_1;
```
